### <span style="color: green;">Questions2: What are correlations between the hobby of villagers and their gender? </span>

#### Explore how many male and female villagers like fitness

Which villager loves fitness in their daily life?

```python
fit_df = Vi[Vi.Hobby == 'Fitness']
fit_df
```

In these villagers, which are male, which are female?

```python
male_fit_df = Vi[(Vi.Hobby == 'Fitness') & (Vi.Gender == 'Male')]
male_fit_df
```

```python
female_fit_df = Vi[(Vi.Hobby == 'Fitness') & (Vi.Gender == 'Male')]
female_fit_df
```

How many male villagers like fitness? How many female villagers like fitness?

```python
gender_fit = fit_df.groupby('Gender').size()
gender_fit
```

Visualize this data with a pie chart by the percentage.

```python
colors = ['green', 'silver']
gender_fit.plot (kind = 'pie', title = 'Gender Distribution of Villagers Whose Hobby is Fitness of AC:NH',labels=gender_fit.index, autopct='%1.1f%%',colors = colors)
```

#### Expolre the gender distribution in all hobbies

Count the hobby types, this is to look at how many villagers of each hobby.

```python
Vi['Hobby'].value_counts()
```

Count the gender numbers in each kind of hobby, this is to filter the data needed.

```python
counts_ho =Vi.groupby('Hobby')['Gender'].value_counts()

counts_ho
```

Count the percentage of gender numbers in each hobby, this is to analyze the gender distribution in each hobby through numbers.

```python
percentage_counts_ho = Vi.groupby('Hobby')['Gender'].value_counts(normalize=True) * 100

percentage_counts_ho
```

Merge the two series above. This can generate combination of the two results above and perceive the gender distribution in each hobby more comprehensively.

```python

ho_gender = pd.concat([counts_ho.rename('Count'), percentage_counts_ho.rename('Percentage')], axis=1)

ho_gender
```

Compare the count of males and females in each hobby category. Visualize it with a bar chart.Directly show the gender distribution of each hobby.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))

color_palette = {"Female": "#34A853", "Male": "#4285F4"}  

counts_ho_df = counts_ho.reset_index(name='Count')
barplot = sns.barplot(x='Hobby', y='Count', hue='Gender', data=counts_ho_df, palette=color_palette)


plt.title('Bar Chart of Hobby Participation by Gender', fontsize=16)
plt.xlabel('Hobby', fontsize=14)
plt.ylabel('Count', fontsize=14)


plt.legend(title='Gender', title_fontsize='13', loc='upper left', fontsize='12')


```

Try another visualization method. Visualize the participants in each hobby by gender with a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt


counts_ho_df = counts_ho.reset_index(name='Count')
counts_pivot = counts_ho_df.pivot('Hobby', 'Gender', 'Count')


plt.figure(figsize=(10, 8))
sns.heatmap(counts_pivot, annot=True, cmap="YlGnBu", fmt=".1f")
plt.title('Heat Map of Hobby Participation by Gender ')
plt.ylabel('Hobby')
plt.xlabel('Gender')
plt.tight_layout()
plt.show()
```
